```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
import logging

headers = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "en-US,en;q=0.9",
    "Cookie": "zero-chakra-ui-color-mode=light-zero; AMP_MKTG_8f1ede8e9c=JTdCJTdE; AMP_8f1ede8e9c=JTdCJTIyZGV2aWNlSWQlMjIlM0ElMjIwZjdiZTMzYS1jZTI5LTQwNzItYmYwOC01ZjFjYTQyMjY1ZjYlMjIlMkMlMjJ1c2VySWQlMjIlM0ElMjIyZTU3N2YwOC04NGMyLTRkYzUtOWFjYy1mNWVjNDJmYjIxNTQlMjIlMkMlMjJzZXNzaW9uSWQlMjIlM0ExNzE0Mjk3NzE4MDM0JTJDJTIyb3B0T3V0JTIyJTNBZmFsc2UlMkMlMjJsYXN0RXZlbnRUaW1lJTIyJTNBMTcxNDI5NzcxODE4MyUyQyUyMmxhc3RFdmVudElkJTIyJTNBMyU3RA==",
    "Priority": "u=0, i",
    "Sec-Ch-Ua": "\"Chromium\";v=\"124\", \"Google Chrome\";v=\"124\", \"Not-A.Brand\";v=\"99\"",
    "Sec-Ch-Ua-Mobile": "?0",
    "Sec-Ch-Ua-Platform": "\"Windows\"",
    "Sec-Fetch-Dest": "document",
    "Sec-Fetch-Mode": "navigate",
    "Sec-Fetch-Site": "cross-site",
    "Sec-Fetch-User": "?1",
    "Upgrade-Insecure-Requests": "1",
    "Referer": "https://www.google.com/",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
    "X-Amzn-Trace-Id": "Root=1-662e1b7d-6ce39ea510b9f2da513a9fd2"
  }

URLS = [
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Waterproof-Remover-2x400ml/product-reviews/B08KQSM846/ref=cm_cr_arp_d_paging_btm_10?ie=UTF8&pageNumber=10&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hyaluronic-Cellular-Refilling-Anti-Wrinkle/product-reviews/B085SD14N4/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/Anti-wrinkle-revitalizing-anti-aging-avocado-panthenol/product-reviews/B078BSRMR4/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/Anti-wrinkle-revitalizing-anti-aging-avocado-panthenol/product-reviews/B078BSRMR4/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/Anti-wrinkle-revitalizing-anti-aging-avocado-panthenol/product-reviews/B078BSRMR4/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/Anti-wrinkle-revitalizing-anti-aging-avocado-panthenol/product-reviews/B078BSRMR4/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Shimmer-Radiant-Care-0-17/product-reviews/B00GKTSQGA/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_10?ie=UTF8&pageNumber=10&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Face-Cream-Moisturise-Light/product-reviews/B07BXRLMY1/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Face-Cream-Moisturise-Light/product-reviews/B07BXRLMY1/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hydrating-Organic-Water-Types/product-reviews/B08S53S64W/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Hydrating-Organic-Water-Types/product-reviews/B08S53S64W/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Toner-Hydrating-Organic-Water/product-reviews/B08S516X6Y/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Toner-Hydrating-Organic-Water/product-reviews/B08S516X6Y/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Strips-Refining-Clear-Up-Citrid/product-reviews/B000ORWFI4/ref=cm_cr_arp_d_paging_btm_10?ie=UTF8&pageNumber=10&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Scrub-Urban-Claywash-Matify/product-reviews/B07VYXWBTH/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Scrub-Urban-Claywash-Matify/product-reviews/B07VYXWBTH/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-LUMINOUS-EVEN-GLOW-Bundle/product-reviews/B08VNZTM28/ref=cm_cr_arp_d_paging_btm_10?ie=UTF8&pageNumber=10&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_10?ie=UTF8&pageNumber=10&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Glowing-Organic-Raspberry-Sensitive/product-reviews/B07ZYCFFQD/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Glowing-Organic-Raspberry-Sensitive/product-reviews/B07ZYCFFQD/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Glowing-Organic-Raspberry-Sensitive/product-reviews/B07ZYCFFQD/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Glowing-Organic-Raspberry-Sensitive/product-reviews/B07ZYCFFQD/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Glowing-Organic-Raspberry-Sensitive/product-reviews/B07ZYCFFQD/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Glowing-Organic-Raspberry-Sensitive/product-reviews/B07ZYCFFQD/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Glowing-Organic-Raspberry-Sensitive/product-reviews/B07ZYCFFQD/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Exfoliating-Natural-Carnitin-Vitamin/product-reviews/B07MVYFXL9/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Exfoliating-Natural-Carnitin-Vitamin/product-reviews/B07MVYFXL9/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Exfoliating-Natural-Carnitin-Vitamin/product-reviews/B07MVYFXL9/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Sheet-Radiance-Ubran-Butter/product-reviews/B08KR28RRB/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Sheet-Radiance-Ubran-Butter/product-reviews/B08KR28RRB/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Sheet-Radiance-Ubran-Butter/product-reviews/B08KR28RRB/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Sheet-Radiance-Ubran-Butter/product-reviews/B08KR28RRB/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Wipes-Micellar-Organic-Water/product-reviews/B08S5254PR/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6?ie=UTF8&pageNumber=1",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9",
    "https://www.amazon.ae/NIVEA-Micellar-Organic-Types-150ml/product-reviews/B07NGTPPM6/ref=cm_cr_arp_d_paging_btm_10?ie=UTF8&pageNumber=10",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber=2&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_3?ie=UTF8&pageNumber=3&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_4?ie=UTF8&pageNumber=4&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_5?ie=UTF8&pageNumber=5&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_6?ie=UTF8&pageNumber=6&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_7?ie=UTF8&pageNumber=7&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_8?ie=UTF8&pageNumber=8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_9?ie=UTF8&pageNumber=9&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Remover-Organic-2x400ml/product-reviews/B08KQZN1SD/ref=cm_cr_arp_d_paging_btm_10?ie=UTF8&pageNumber=10&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Moisturizer-Anti-Wrinkle-Energy-Vitamin/product-reviews/B07VVPLFLW/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Cream-Moisturizer-Natural-SPF30/product-reviews/B07D68T3MQ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Under-Eye-Hydrating-Organic-Hyaluron/product-reviews/B08X9XMRY4/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/Micellar-Expert-Waterproof-Cleansing-Mousse/product-reviews/B081VB59RV/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.ae/NIVEA-Micellar-Naturally-Organic-Normal/product-reviews/B09CPB7SLQ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
    ]



def get_page_html_and_product_name(page_url: str) -> tuple:
    resp = requests.get(page_url, headers=headers)
    soup = BeautifulSoup(resp.text, "lxml")
    product_name = soup.find('a', {'class': 'a-link-normal'}).get_text().strip() if soup.find('a', {'class': 'a-link-normal'}) else 'Unknown Product'
    return resp.text, product_name


def get_reviews_from_html(page_html: str) -> BeautifulSoup:
    soup = BeautifulSoup(page_html, "lxml")
    reviews = soup.find_all("div", {"class": "a-section celwidget"})
    return reviews


def get_review_date(soup_object: BeautifulSoup):
    date_string = soup_object.find("span", {"class": "review-date"}).get_text()
    return date_string


def get_review_text(soup_object: BeautifulSoup) -> str:
    review_text = soup_object.find(
        "span", {"class": "a-size-base review-text review-text-content"}
    ).get_text()
    return review_text.strip()


def get_review_header(soup_object: BeautifulSoup) -> str:
    review_header = soup_object.find(
        "a",
        {
            "class": "a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"
        },
    ).get_text()
    return review_header.strip()[19:]


def get_number_stars(soup_object: BeautifulSoup) -> str:
    stars = soup_object.find("span", {"class": "a-icon-alt"}).get_text()[:4]
    return stars.strip()



def orchestrate_data_gathering(single_review: BeautifulSoup, product_name) -> dict:
    return {
        "product_name" : product_name,
        "review_title": get_review_header(single_review),
        "review_date": get_review_date(single_review),
        "review_stars": get_number_stars(single_review),
        "review_text": get_review_text(single_review)
    }


if __name__ == '__main__':
    logging.basicConfig(level=logging.INFO)
    all_results = []
    for u in URLS:
      print(all_results)
      logging.info(u)
      product_name = get_page_html_and_product_name(u)[1]
      if (product_name == "Unknown Product"):
          product_name = all_results[-1]["product_name"]
      print(product_name)
      html = get_page_html_and_product_name(u)[0]
      reviews = get_reviews_from_html(html)
      for rev in reviews:
        try:
          data = orchestrate_data_gathering(rev, product_name)
          all_results.append(data)
        except:
          pass

    out = pd.DataFrame.from_records(all_results)
    logging.info(f"{out.shape[0]} Is the shape of the dataframe")
    save_name = f"{datetime.now().strftime('%Y-%m-%d-%m')}.csv"
    logging.info(f"saving to {save_name}")
    out.to_csv(save_name)
    logging.info('Done yayy')
```
